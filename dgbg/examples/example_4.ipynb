# Example to highlight integration with AutoDock Vina.

```python
from dgbg import dgbg
import numpy as np
from vina import Vina
```

```python
box = dgbg.binding_box('pdb_files/7uty.pdb')
```

```python
ligand_data = box.get_ligand_data('OFR', 'A')
```

```python
ligand_data
```

```python
centroid, box_size = box.create_box(spacing=0.375, padding=np.array([10, 5, 5]))
```

```python
box_size
```

```python
centroid
```

```python
result = box.show_result(centroid, box_size)
```

```python
result
```

```python
v = Vina(sf_name='vina')
```

```python
v.set_receptor('pdbqt_files/7uty.pdbqt')
```

```python
v.set_ligand_from_file('pdbqt_files/ligand_7uty.pdbqt')
```

```python
v.compute_vina_maps(center=centroid, box_size=box_size)

# Score the current pose
energy = v.score()
print('Score before minimization: %.3f (kcal/mol)' % energy[0])

# Minimized locally the current pose
energy_minimized = v.optimize()
print('Score after minimization : %.3f (kcal/mol)' % energy_minimized[0])
v.write_pose('pdbqt_files/7uty_ligand_vina_minimized.pdbqt', overwrite=True)

# Dock the ligand
v.dock(exhaustiveness=8, n_poses=5)
v.write_poses('pdbqt_files/7uty_ligand_vina_out.pdbqt', n_poses=5, overwrite=True)
```
